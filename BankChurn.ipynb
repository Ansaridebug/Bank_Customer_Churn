```python
data = data.drop(['CustomerId', 'Surname'],axis=1)
```

```python
data.head()
```

```python
# Encoding categorical data (one-hot)
data['Geography'].unique()
```

```python
# Convert this geograpy column values into binary vector
# F S G
# 0 0 1
# 0 1 0
# 1 0 0
# To create this dummy variable we use get_dummy()
data = pd.get_dummies(data,drop_first=True,dtype=int)
```

```python
x = data.drop('Exited',axis=1)
y = data['Exited']
```

```python
# Handling imbalanced data with SMOTE
# SMOTE: Synthetic Minority Oversampling Technique
```

```python
from imblearn.over_sampling import SMOTE
```

```python
x_res,y_res = SMOTE().fit_resample(x,y)
```

```python
y_res.value_counts()
```

```python
# Splitting the dataset into the Training Set and Test set
from sklearn.model_selection import train_test_split
```

```python
# x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.20,random_state=42,stratify=y)
x_train,x_test,y_train,y_test = train_test_split(x_res,y_res,test_size=0.20,random_state=42)
```

```python
# Feature scaling
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
x_train
```

```python
# Logistic Regression
from sklearn.linear_model import LogisticRegression
```

```python
log = LogisticRegression()
```

```python
log.fit(x_train,y_train)
```

```python
y_pred = log.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred) # 0.808
```

```python
accuracy_score(y_test,y_pred)
```

```python
# Our dataset is imbalance, it is very dangerous to use accuracy as a metrics on imbalance datasets
# So, for imbalance dataset we have to check precision_score,recall_score,f1_score.
from sklearn.metrics import precision_score,recall_score,f1_score
```

```python
# It represent the models ability to correctly predict the + out of all the + predictions it mades
precision_score(y_test,y_pred) # 0.5891
```

```python
precision_score(y_test,y_pred)
```

```python
# It represent the models ability to correctly predict the + out of actual +
recall_score(y_test,y_pred) # 0.1867
```

```python
recall_score(y_test,y_pred)
```

```python
# It represent the models score as a function of precision and f1_score
f1_score(y_test,y_pred) # 0.2835
```

```python
f1_score(y_test,y_pred)
```
